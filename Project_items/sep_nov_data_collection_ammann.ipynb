```python
# Combine csv files for top 200 in september

import datetime
import glob
import os
import pandas as pd 

# Merge top 200
files = glob.glob('Raw_files\top_200\sep\*.csv')
# print(files)

sept_combined_df_top200 = pd.concat([pd.read_csv(fp, skiprows=[0], skip_blank_lines=True).assign(Filename=os.path.basename(fp)) for fp in files])

sept_combined_df_top200

# Make new columns for month and week numbers
sept_combined_df_top200["Month"] = ""
sept_combined_df_top200["Week Number"] = ""
sept_combined_df_top200["Type"] = ""

# Extract Date from filename using slice top 200 slice is 19, 29
sept_combined_df_top200["Date"] = sept_combined_df_top200['Filename'].str.slice(19,29)
# convert to Datetime
sept_combined_df_top200["Date"]=pd.to_datetime(sept_combined_df_top200["Date"])
# get week number form Date
sept_combined_df_top200["Week Number"] = sept_combined_df_top200['Date'].dt.week

# Make loop to add values in all rows in datafram
for index, row in sept_combined_df_top200.iterrows():
    sept_combined_df_top200.loc[index, "Month"] = "September"
    sept_combined_df_top200.loc[index, "Type"] = "Top 200"


#save as csv
sept_combined_df_top200.to_csv("csv files\sept_combined_top200.csv")

# sort values by week number and position for chronological order and top songs
sept_combined_df_top20 = sept_combined_df_top200.sort_values(['Week Number', 'Position'], ascending=True)
sept_combined_df_top20


```

```python
# Use the URL to look up artist info to pull genre information

# Import dependencies

import json
import pandas as pd
import time

#open csv file and read with pandas JIC
csv_path = "csv files\\sept_nov_viral_combined.csv"
sep_nov_viral_df = pd.read_csv(csv_path)
sep_nov_viral_df.head()

# Make blank column for storing the artist ID
sep_nov_viral_df['Artist ID'] = ""
sep_nov_viral_df['Genre 1'] = ""
sep_nov_viral_df['Genre 2'] = ""
sep_nov_viral_df['Genre 3'] = ""

# Look up and store artist ID using the track URL and spotipy API
for index, row in sep_nov_viral_df.iterrows():

    track_id = row['URL']
    track_features = sp.track(track_id)
    sep_nov_viral_df.loc[index, 'Artist ID'] = track_features['album']['artists'][0]['id']

# Take the artist ID and gather genre information using spotipy API
for index, row in sep_nov_viral_df.iterrows():

    genre_id = row['Artist ID']
    artist_features = sp.artist(genre_id)
    # Some songs didn't have a genre and some have multiple, pulling the first 3 and printing skip for blanks
    try:
        sep_nov_viral_df.loc[index, 'Genre 1'] = artist_features['genres'][0]
        sep_nov_viral_df.loc[index, 'Genre 2'] = artist_features['genres'][1]
        sep_nov_viral_df.loc[index, 'Genre 3'] = artist_features['genres'][2]
    except (IndexError, KeyError):
        print(f'skipped')
    # Add sleep timer to not get kicked off the API while pulling a lot of data
    time.sleep(1.1)

sep_nov_viral_df.head(25)
```
